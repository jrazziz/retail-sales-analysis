# Retail Sales Analysis: Key Insights

## Executive Summary

This analysis of 10,194 transactions (2023–2026) reveals three critical findings:

1. **Heavy discounts destroy profitability.** Orders with 50%+ discounts lose 118% margin — 
   the company pays customers to take products.

2. **Q4 (Oct–Nov) drives 40% of annual revenue**, but January–February are consistently weak 
   and sometimes unprofitable.

3. **Corporate customers generate the highest profit per customer.** The top 10 customers are 
   mostly Corporate buyers of copiers and binding systems.

## Finding 1: Discounts Above 20% Erode Profitability

**Evidence:** From SQL analysis (`03_discount_impact.sql`) and Chart 2.

| Discount Tier | Margin | Status |
|-------------|--------|--------|
| No Discount | 29.6% | ✅ Excellent |
| Low (1-20%) | 12.7% | ⚠️ Acceptable |
| Medium (21-50%) | 3.0% | ❌ Dangerous |
| High (>50%) | -118.7% | 🚨 Catastrophic |

**Key Insight:** 42% of all orders use medium-to-high discounts and barely break even or lose money. 
The same products (binding systems, printers) generate 40%+ margin at no discount.

**Recommendation:** Implement a strict discount approval policy. Cap discounts at 20% without 
manager approval. Eliminate discounts above 50% entirely.

## Finding 2: Strong Q4 Seasonality, Weak Q1

**Evidence:** From time trend analysis (`05_time_trends.sql`) and Chart 3.

- **Peak months:** November ($79K-$118K), December ($72K-$97K)
- **Trough months:** February ($4K-$22K), January (sometimes unprofitable)
- **Growth trend:** 2026 outperformed all previous years

**Key Insight:** The business is growing but dangerously dependent on Q4. January 2024 lost $3,190 
despite $18K in sales — likely due to post-holiday discounting of leftover inventory.

**Recommendation:**
- Build inventory forecasting to reduce January clearance discounts
- Launch Q1 marketing campaigns to smooth revenue curve
- Hire seasonal staff for Q4, reduce Q1 overhead

## Finding 3: Corporate Segment is the Profit Engine

**Evidence:** From customer analysis (`04_customer_analysis.sql`) and Chart 5.

| Metric | Corporate | Consumer | Home Office |
|--------|-----------|----------|-------------|
| Avg Profit per Customer | Highest | Medium | Lower |
| Top 10 Presence | 5-6 | 3-4 | 1-2 |
| Typical Purchase | Copiers, binding systems | Paper, labels | Occasional equipment |

**Key Insight:** Corporate buyers purchase high-margin equipment with minimal discounting. 
Tamara Chand (Corporate) generated $8,981 profit at 47% margin. Meanwhile, high-discount 
Consumer orders on furniture lose money.

**Recommendation:**
- Assign dedicated sales reps to top 20 Corporate customers
- Create Corporate loyalty program with volume incentives (not discounts)
- Reduce furniture inventory for Consumer segment — it consistently loses money

## Conclusion

This analysis reveals a profitable business with critical vulnerabilities:

| Strength | Weakness |
|----------|----------|
| Growing revenue | Over-reliance on Q4 |
| Strong Corporate segment | Destructive discounting culture |
| High-margin consumables | Loss-making furniture category |

**Priority Actions:**
1. 🔴 **Immediate:** Cap discounts at 20%, eliminate >50% discounts
2. 🟡 **Short-term:** Reduce furniture inventory, focus on Office Supplies + Technology
3. 🟢 **Long-term:** Build Q1 revenue streams, expand Corporate sales team
